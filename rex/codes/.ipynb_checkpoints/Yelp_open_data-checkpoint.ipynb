# Yelp Dataset (Beta)

```python
import json
import pandas as pd
import numpy as np
```

```python
with open('./datas/yelp_academic_dataset_business.json', 'r') as f:
    data = [json.loads(x) for x in f.readlines()]
```

## Import Data

```python
price = [i['attributes'] if i['attributes'] != None else {'RestaurantsPriceRange2':np.NAN} for i in data]
```

```python
price = pd.DataFrame(price)
```

```python
price.head()
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
final_df = pd.concat([df, price], axis=1)
```

```python
# final_df = final_df[final_df['city']=='Madison']
```

```python
final_df = final_df[['business_id','categories','city','latitude','longitude','name','neighborhood','postal_code','state','RestaurantsPriceRange2']]
```

```python
final_df.head()
```

What happened ?
 - I pulled two data from our datasets. One named "price" and the other one named "data", Since not all data in the price dataframe has the target value which is the RestaurantsPriceRange2, therefore, I assigned all the buseinsses that doesn't have RestaurantsPriceRange2 to Nan to be able to convert into Dataframe to do futher data cleaning. 
 - I, then filtered out (business_id, categories, city, latitude, longitude, name, neighborhood, postal_code, state, RestaurantsPriceRange2) into on DataFrame to move to the next step.

## Data Cleaning

```python
final_df.isnull().sum()
```

```python
final_df = final_df.dropna(subset=['RestaurantsPriceRange2'])
final_df.reset_index(drop=True, inplace=True)
```

```python
final_df.head()
```

```python
final_df['price'] = pd.to_numeric(final_df['RestaurantsPriceRange2'])
final_df.drop(['RestaurantsPriceRange2'], axis=1, inplace=True)
final_df['postal_code'] = pd.to_numeric(final_df.postal_code, errors='coerce').fillna(0).astype(np.int64)
```

```python
final_df.info()
```

## Zillow Boundaries to DataFrame

```python
import geopandas as gpd
from shapely.geometry import Point, Polygon
```

```python
zillow_boundaries = gpd.read_file('./datas/zillow-neighborhoods.geojson')
```

```python
zillow_boundaries.head()
```

```python
zillow_boundaries['geometry'][40]
```

```python
final_df['geometry'] = list(zip(final_df.longitude, final_df.latitude))
final_df['geometry'] = final_df['geometry'].apply(lambda x: Point(x)) 
```

```python
zillow_geo = gpd.GeoDataFrame(zillow_boundaries[['regionid','geometry']])
yelp_geo = gpd.GeoDataFrame(final_df[['business_id','geometry']]) 
```

```python
zillow_geo.dropna(inplace=True)
```

```python
zillow_geo.head()
```

```python
yelp_geo.head()
```

```python
merged_geos = gpd.sjoin(yelp_geo, zillow_geo, op='within', how='left').fillna('NoNeighborhood')
```

```python
merged_df = pd.DataFrame(merged_geos[['business_id','regionid']])
```

```python
merged_df.shape
```

```python
dataset_yelpopen = final_df.merge(merged_df, on='business_id', how='left')
```

```python
dataset_yelpopen.head()
```

```python
zillow_stats = pd.read_csv('./datas/Neighborhood_PriceToRentRatio_AllHomes.csv')
```

```python
zillow_stats.columns = [i.lower() for i in zillow_stats.columns]
```

```python
zillow_stats.head()
```

```python
tidy_stats = zillow_stats[['regionid','2018-11']]
```

```python
tidy_stats.columns = ['regionid','price_to_rent']
```

```python
dataset_yelpopen['regionid'] = dataset_yelpopen['regionid'].astype(str)
tidy_stats['regionid'] = tidy_stats['regionid'].astype(str)
```

```python
model_df = dataset_yelpopen.merge(tidy_stats,on='regionid',how='left').fillna('no_zillow_data')
```

```python
#Drop items
model_df.drop(['business_id','latitude','longitude'], axis=1, inplace=True)
# check the final df
model_df.head()
```

- I saved import Zillow map that were based on the latitude and longitude, so saved from the origional dataset, I merged based on that. Also, worked with Zip_code
