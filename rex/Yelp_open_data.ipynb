```python
import json
import pandas as pd
import numpy as np
```

```python
with open('./datas/yelp_academic_dataset_business.json', 'r') as f:
    data = [json.loads(x) for x in f.readlines()]
```

```python
price = [i['attributes'] if i['attributes'] != None else {'RestaurantsPriceRange2':np.NAN} for i in data]
```

```python
price = pd.DataFrame(price)
```

```python
df = pd.DataFrame(data)
```

```python
final_df = pd.concat([df, price], axis=1)
```

```python
final_df = final_df[final_df['city']=='Madison']
```

```python
final_df = final_df[['business_id','categories','city','latitude','longitude','name','neighborhood','postal_code','state','RestaurantsPriceRange2']]
```

```python
final_df.isnull().sum()
```

```python
final_df = final_df.dropna(subset=['RestaurantsPriceRange2'])
final_df.reset_index(drop=True, inplace=True)
```
